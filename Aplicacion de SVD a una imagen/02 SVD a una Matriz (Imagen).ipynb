```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')
imagen = Image.open("..\imagenes\imagen_1.jpg")
plt.imshow(imagen)
```

```python
imagen_gr = imagen.convert('LA')#Pasa la imagen a escala de grises
```

```python
plt.imshow(imagen_gr)
```

```python
imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)
imagen_mat.shape
```

```python
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])#Ajustar las dimensiones de la imagen
```

```python
imagen_mat.shape
```

```python
U, D, V = np.linalg.svd(imagen_mat)
```

```python
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
imagen_recons = np.matrix(U[:,:1])* np.diag(D[:1])*np.matrix(V[:1,:])
plt.imshow(imagen_recons, cmap='gray')
```

```python
i = 5
imagen_recons = np.matrix(U[:,:i])* np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')

title = "valores singulares = %s" % i
plt.title(title)
plt.show()
```
